# comparison by gender

```python
columns_to_plot = ['Quality of Sleep', 'Sleep Duration', 'Stress Level']


for column in columns_to_plot:
    plt.figure(figsize=(8, 6))
    sns.barplot(data=df, x='Gender', y=column)
    plt.title(f'{column} by Gender')
    plt.xlabel('Gender')
    plt.ylabel(column)
    plt.show()
```

comparison by age

```python
columns_to_plot = ['Quality of Sleep', 'Sleep Duration', 'Stress Level']
for column in columns_to_plot:
  plt.figure(figsize=(8, 6))
  sns.lineplot(x='Age', y=column, data=df)  
  plt.title(f'{column} by Age')
  plt.xlabel('Age')
  plt.ylabel(column)
  plt.show()
```

# comparison of occupation and bmi category

```python
cross_tab = pd.crosstab(df['Occupation'], df['BMI Category'])

# Plotting a stacked bar chart
plt.figure(figsize=(8, 6))
cross_tab.plot(kind='bar', stacked=True)
plt.title('Comparison of Occupation and BMI Category')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='BMI Category')
plt.show()
```
